```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Solving a System of Linear Equations

A = np.array([[27, -10, 4, -29], [-16, 5, -2, 18], [-17, 4, -2, 20], [-7, 2, 1, 8]])
B = np.array([[1], [-1], [0], [1]])
result = np.linalg.solve(A, B)
print(result)
```

```python
# checking my answer with transformation

check = A @ result
print(check)
```

```python
# 2. Visualizing a Linear Transformation
# define matrix A

A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2], [np.sqrt(2)/2, np.sqrt(2)/2]])
print(A)
```

```python
# 2a. Find T(x), where x = [2, -2]

v = np.array([2, -2])
print(A @ v)
```

```python
# 2b. Mimic the code in Section 7.1.3

x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)
# show plot
plt.scatter(xx, yy, s = 20, c = xx + yy)
plt.show()
```

```python
# define transformation T

T = np.array(A)
# shape (x,y) values
xy = np.vstack([xx.flatten(), yy.flatten()])
# show the shape
print(xy.shape)
```

```python
# evaluate transformed coordinates
trans = T @ xy
# show the shape
print(trans.shape)
```

```python
# create two arrays with similar shape to xx and yy

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)
```

```python
# plot orginal and transformed grid

f, axes = plt.subplots(1, 2, figsize = (6, 3))
axes[0].scatter(xx, yy, s = 10, c = xx + yy)
axes[1].scatter(xx_transformed, yy_transformed, s = 10, c = xx + yy)
plt.show()
print("T (the defined linear transformation) rotates the coordinates by 45 counterclockwise.")
```
